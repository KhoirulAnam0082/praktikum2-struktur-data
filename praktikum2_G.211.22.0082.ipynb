Dictionary

```python
# Membuat dictionary
data_mahasiswa = {
    "nim": "12345",
    "nama": "John Doe",
    "jurusan": "Informatika",
    "ipk": 3.75
}

# Mengakses nilai dalam dictionary
print("NIM:", data_mahasiswa["nim"])
print("Nama:", data_mahasiswa["nama"])
print("Jurusan:", data_mahasiswa["jurusan"])
print("IPK:", data_mahasiswa["ipk"])

# Mengubah nilai dalam dictionary
data_mahasiswa["ipk"] = 3.9
print("IPK terbaru:", data_mahasiswa["ipk"])

# Menambahkan item baru ke dictionary
data_mahasiswa["semester"] = 5
print("Semester:", data_mahasiswa["semester"])

# Iterasi melalui keys dalam dictionary
print("Data mahasiswa:")
for key in data_mahasiswa:
    print(key, ":", data_mahasiswa[key])
```

Tuple


```python
# Membuat tuple
buah_tuple = ("apel", "jeruk", "pisang", "anggur")

# Mengakses elemen tuple
print("Elemen pertama:", buah_tuple[0])
print("Elemen terakhir:", buah_tuple[-1])

# Iterasi melalui tuple
print("Semua buah:")
for buah in buah_tuple:
    print(buah)

# Menentukan panjang tuple
panjang_tuple = len(buah_tuple)
print("Jumlah buah dalam tuple:", panjang_tuple)
```
